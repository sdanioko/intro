### 6-  Define a method iterate(self, z), where z is a complex number, which runs the following algorithm:
### i. initialize a counter m to 0
### ii. apply the method juliamap to z, increment the counter m
### iii. if |z|> 2, return the value of the counter m
### iii. otherwise, if m >= n, return the value 0
### iv. otherwise, go back to step ii.

```python
import math 

class JuliaSet:
    def __init__(self, c, m = 100):
        self._d = 0.001
        self.c = c
        self.m = m

    def juliamap(self,z):
        return z**2 + self.c
    
    def iterate(self,z):
        for counter in range(1,self.m+1): # Normally, this counter should start at zero. But, if we 
            # start zero, we will never know we went through the loop once
            z = self.juliamap(z)
            if math.sqrt(z.real**2 + z.imag**2) > 2 :
                return counter
        return 0
                
    
```
